## 1. Introduction

**In this notebook, we create a classifer to predict house prices in 1987s Boston (Ma.).**
The classifier takes input data which corresponds to a *longitude* and *latitude* float64 tuple, therefore aiming to establish a non-trivial mapping between geographical location and house price.

In the provided dataset; the corrected median house price is supplied to us under the `CMEDV` key.
This value is measured in 1000s of USD and characterises the average house price for a specific census location, described under the `TOWN` and `TOWNNO` keys, but more accurately under the `LON` & `LAT` keys, which correspond to longitude and latitude.

The main classifier architecture will use the two main input features: `LON` and `LAT` to both:
* generate sub inputs via calls to nested classifiers with location based inputs which produce engineered features (we will cover this in more depth later)
* predict a novel `CMEDV`

The final classifier will seek a mapping $F_f$ for a novel test point $(x_{lon}, x_{lat})$, given by:

$$F_f[(x_{lon}, x_{lat}), F_{nested}[(x_{lon}, x_{lat})], \dots] \to \text{CMEDV_novel}$$
